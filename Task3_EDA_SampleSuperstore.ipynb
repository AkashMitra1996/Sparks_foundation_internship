# Proper Visualization of the Data set

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=df,color='green')
plt.show()
```

## Quantites ordered by state

```python
plt.figure(figsize = (20, 9))
sns.countplot(x = df['State'], palette = 'winter_r', order = (df['State'].value_counts().head(50)).index)
plt.xticks(rotation = 50, ha = 'right', fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('State', fontsize = 15)
plt.ylabel('Quantities ordered', fontsize = 15)
```

## Sub-category Wise Distribution In Different Regions

```python
plt.figure(figsize = (30, 12))
sns.countplot(x = 'Sub-Category', hue = 'Region', data = df,palette='twilight_r')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Sub-Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

## Visualizing sub-category wise distribution of sales

```python
# visualizing sub-category wise distribution of sales
plt.figure(figsize=(12,10))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

## Customer's per segment

```python
#What is the distribution of our customer segment
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
p = sns.countplot(x="Segment", data=df, ax=ax)
ax.set_title("Customer Distribution by Segment")
ax.set_xticklabels(p.get_xticklabels(), rotation=90)
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+1))
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x=df['Segment'])
```

## Categories classified according to the region.

```python

plt.figure(figsize = (30, 12))
sns.countplot(x = 'Category', hue = 'Region', data = df, palette = 'icefire')
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlabel('Categories', fontsize = 25)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20)
```

## Sales by Region

```python
sales_by_region = df.groupby('Region').size()
plt.figure(figsize=(10,6))
plot_by_region = sales_by_region.plot(title='Sales by Region',xticks=(1,2,3,4),lw=3,c='blue')
plot_by_region.set_xlabel('Region',size=12)
plot_by_region.set_ylabel('Sales',size=12)
```

## Observation:

- South Region has worst performance 

- East & West have beter performance 

## Sales by Ship Mode

```python
sales_by_shipmode = df.groupby('Ship Mode').size()
print(sales_by_shipmode)
plt.figure(figsize=(10,6))
sales_by_shipmode = sales_by_shipmode.plot(title='Sales by Ship Mode',lw=3,c='g')
sales_by_shipmode.set_xlabel('Ship Mode',size=12)
sales_by_shipmode.set_ylabel('Sales',size=12)
```

## Observations:

- Standard Class Shipments is choosen by Majority of customers

- Many customers opt for Second Class Shipment

- Customers opting for Same Day Shipment a very less
